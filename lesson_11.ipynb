```python
import os
import glob
import numpy as np
def read_sequence_from_fasta(filename='data/salmonella_spi1_region.fna'):

    if not os.path.isfile(filename):
        raise RuntimeError('File does not exist.')
    
    metadata = ''
    seq = ''
    with open(filename, 'r') as f:
        metadata = f.readline().strip()
        if not metadata[0] == '>':
            seq = metadata
        for line in f:
            seq += line.strip()
        
    return metadata, seq
```
